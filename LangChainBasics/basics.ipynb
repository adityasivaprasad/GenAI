```python
import langchain
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["GROQ_API_KEY"]=os.getenv("NEW_GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage, SystemMessage
```

```python
model=init_chat_model("groq:llama-3.1-8b-instant")
```

```python
messages = [SystemMessage("You are a helpful assistant."), HumanMessage("Write me a Python program to calculate if a number is pallindrome ?")]
response = model.invoke(messages)
print(response.content)
```

## Streaming Output Messages

```python
for chunk in model.stream(messages):
    print(chunk.content, end="", flush=True)
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are an efficient translator. Translate the following {source_text} from {source_lang} to {target_lang}"),
    ("human", "{source_text}")
])
```

```python
prompt = prompt_template.invoke({
    "source_text": "What is machine learning and its advantages ?",
    "source_lang": "English",
    "target_lang": "Hindi"
})
```

```python
prompt
```

```python
response = model.invoke(prompt)
print(response.content)
```
