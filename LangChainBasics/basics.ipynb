```python
import langchain
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["GROQ_API_KEY"]=os.getenv("NEW_GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage, SystemMessage
```

```python
model=init_chat_model("groq:llama-3.1-8b-instant")
```

```python
messages = [SystemMessage("You are a helpful assistant."), HumanMessage("Write me a Python program to calculate if a number is pallindrome ?")]
response = model.invoke(messages)
print(response.content)
```
